Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
x=["Alex","Jon","Ben","Sam","Smith"]
y=[90,50,100,70,93]
colors = ['red', 'green', 'blue', 'yellow','olive']
plt.bar(x,y,color=colors,edgecolor="black")
plt.title("Student Marks Distribution",fontsize=20)
plt.xlabel("Student Names")
plt.ylabel("Marks out of 100")
plt.show()
```

Line Chart

```python
x=["Alex","Jon","Ben","Sam","Smith"]
y=[90,50,100,70,93]
plt.plot(x,y,marker="o",ls="--")
plt.title("Student Marks Distribution")
plt.xlabel("Student Names")
plt.ylabel("Marks out of 100")
plt.show()
```

```python
x=["Alex","Jon","Ben","Sam","Smith"]
y=[90,50,100,70,93]
y2=[76,89,70,98,84]
y3=[67,94,80,58,99]
plt.plot(x,y,marker="o",label="test1")
plt.plot(x,y2,marker="o",ls="--",label="test2")
plt.plot(x,y3,marker="o",ls="-",label="test3",color="green")
plt.title("Student Marks Distribution")
plt.xlabel("Student Names")
plt.ylabel("Marks out of 100")
plt.legend()
plt.show()
```

Pie Chart


```python
import matplotlib.pyplot as plt
brands=["Apple","Samsung","Oneplus","Google","MI"]
x=[35,25,10,17,13]
ex=[0,0.1,0,0,0]
plt.figure(figsize=(15,7))
plt.pie(x,labels=brands,explode=ex,autopct="%.f")
plt.show()
```

Stack Plot

```python
Subjects=[1,2,3,4,5,6,7]

A1=[23,45,50,34,39,28,38]
A2=[45,25,36,42,29,38,22]
A3=[49,38,46,32,39,45,50]

plt.figure(figsize=(12,6))
plt.stackplot(Subjects,A1,A2,A3,labels=["Week#1","Week#2","Week#3"])
plt.title("Student Attendance Stack of last 3 weeks")
plt.xlabel("Days",fontsize=15)
plt.ylabel("Number Student attend the class",fontsize=15)
plt.legend()
plt.show()
```

Step Plot

```python
Months = ["January", "February", "March", "April", "May", "June","July", "August", "September", "October", "November", "December"]
Price=["30k","35k","40k","42k","43k","45k","46k","47k","50k","49k","48k","44k"]

plt.figure(figsize=(12,6))
plt.step(Price,Months,marker="o")
plt.title("Price of a Smart Phone in 2025",fontsize=15)
plt.xlabel("Month End Price",fontsize=15)
plt.ylabel("Months",fontsize=15)
plt.show()
```
