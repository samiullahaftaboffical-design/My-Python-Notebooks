Encoding


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
dataset=pd.read_csv(r"F:\DataSets\Dataset.csv")
# dataset.info()
en=dataset[["Gender"]]
si=OneHotEncoder(drop="first")
x=si.fit_transform(en).toarray()
feature_names=si.get_feature_names_out(['Gender'])
encoded_df=pd.DataFrame(x,columns=feature_names)
dataset = dataset.drop("Gender", axis=1)
dataset = pd.concat([dataset, encoded_df], axis=1)
dataset.to_csv(r"F:\DataSets\Updated_Dataset.csv", index=False)
dataset.head(50)
```

Ordinal Encoding


```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
k=pd.read_csv(r"F:\DataSets\Dataset.csv")
k["Course"]=k["Course"].fillna(k["Course"].mode()[0])
order=[['IT', 'Computer Science', 'Rob', 'CB', 'SE', 'DS', 'AI', 'Gaming', 'Iot']]
os=OrdinalEncoder(categories=order)
k["Course_Encoded"]=os.fit_transform(k[["Course"]])
k
```
