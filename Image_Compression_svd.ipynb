```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image=mpimg.imread('pngtree-d-smartphone-with-gifts-and-discount-percentage-e-commerce-web-banner-image_3753281.jpg')# To read image and it is also make it deffault numpy array
image.shape# To display shape of the array
```

```python
def display(image,title):
    plt.imshow(image)#To load image
    plt.title(f"{title}")#To add title on the image
    plt.axis('off')#To remove axis
    plt.show()#To show image
```

```python
display(image,"Original image")
```

```python
if image.ndim==3:
    gray=image.mean(axis=2)
else:
    gray=image
plt.imshow(gray)
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(gray, full_matrices=False)
```

```python
# Define compression function
def compress_image(U, S, VT, k):
    S_k = np.diag(S[:k])
    U_k = U[:, :k]
    VT_k = VT[:k, :]
    return np.dot(U_k, np.dot(S_k, VT_k))
k=50
compressed=compress_image(U,S,VT,k)
display(image,"Before")
display(compressed,"After")
```
