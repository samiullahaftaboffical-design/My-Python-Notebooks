```python
import cv2
import matplotlib.pyplot as plt
```

Image Rotation

```python
image=cv2.imread(r"ocean.jpg")
image.shape
```

```python
Resize=cv2.resize(image,(275,183))
image=Resize
image.shape
```

```python
cv2.imshow("Ocean",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
h , w = image.shape[:2]
print(f"Height is {h} and Width is {w}")
```

```python
points = [
    ("Top-Left", 0, 0),
    ("Center", w//2, h//2),
    ("Bottom-Right", w, h)
]

plt.imshow(image)
for name, x, y in points:
    plt.scatter(x, y, c="Black", s=50)
    plt.text(x+10, y+10, name, color="Black", fontsize=8)
plt.show()
```

```python
Center=(h//2,w//2)
M=cv2.getRotationMatrix2D(Center,-32,1.0)
Rotated=cv2.warpAffine(image,M,(w,h))
```

```python
cv2.imshow("Orignal",image)
cv2.imshow("Rotated",Rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Image Flipping


```python
Flip=cv2.flip(image,-1)
```

```python
cv2.imshow("Orignal",image)
cv2.imshow("Fliped",Flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
