Drawing a Line on Image


```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"headphone.jpg")
```

```python
RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB)
plt.grid(True)
plt.xticks(range(0, image.shape[1], 25))
plt.yticks(range(0, image.shape[0], 25)) 

plt.show()
```

```python
cv2.line(RGB,(100,160),(250,160),(0,0,255),4)
cv2.line(RGB,(100,160),(100,12),(0,0,255),4)
cv2.line(RGB,(250,160),(250,12),(0,0,255),4)
cv2.line(RGB,(100,12),(250,12),(0,0,255),4)
```

```python
cv2.imshow("Lined",RGB)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Drawing a Rectangle on Image

```python
image2= cv2.imread(r"Spiderman_2.png")
image2.shape
```

```python
cv2.imshow("SpiderMan",image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
R1=cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(R1)
plt.grid(True)
plt.xticks(range(0,R1.shape[1],25))
plt.yticks(range(0,R1.shape[0],25))
plt.show()
```

```python
cv2.rectangle(image2,(35,12),(190,210),(1,1,1),4)
```

```python
cv2.imshow("Rectangle",image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Drawing a Circle on Image

```python
image3 = cv2.imread(r"ball.png")
```

```python
R2=cv2.cvtColor(image3,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(R2)
plt.grid(True)
plt.xticks(range(0,R2.shape[1],75))
plt.yticks(range(0,R2.shape[0],75))
plt.show()
```

```python
Resize=cv2.resize(image3,(225,225))
```

```python
h,w,c=Resize.shape
```

```python
cv2.circle(Resize,(w//2,h//2),100,(139,0,0),4)
```

```python
cv2.imshow("Cricle",Resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
