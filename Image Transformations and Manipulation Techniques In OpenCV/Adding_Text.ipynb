```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"Spiderman_2.png")
```

```python
RGB=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
RGB.shape
```

```python
plt.imshow(RGB)
plt.grid(True)
plt.xticks(range(0,RGB.shape[1],30))
plt.yticks(range(0,RGB.shape[1],30))
plt.show()
```

```python
cv2.line(image,(170,120),(185,120),(1,1,1),2)
cv2.line(image,(170,120),(175,130),(1,1,1),2)
cv2.line(image,(170,120),(175,110),(1,1,1),2)
cv2.putText(image,"Eye",(186,120),cv2.FONT_HERSHEY_SIMPLEX,0.5,(1,1,1),2)
```

```python
cv2.imshow("Spider Man",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
