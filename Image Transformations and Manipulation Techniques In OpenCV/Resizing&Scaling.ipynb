```python
import cv2
import matplotlib.pyplot as plt
```

How to Resize a Image 

```python
image = cv2.imread(r"headphone.jpg")
cv2.imshow("HEADPHONES",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
Resize=cv2.resize(image,(700,500))
```

```python
cv2.imshow("Orignal",image)
cv2.imshow("Resize",Resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
plt.grid(True)
```
